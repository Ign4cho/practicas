```python
import pandas as pd
```

```python
# Ruta del archivo .xlsx
file_path = r"C:\Users\nacho\Desktop\Facultad\Cuarto Año\practicas\revision_documentos.xlsx"

# Lee el archivo Excel
df = pd.read_excel(file_path)

# Muestra las primeras filas del DataFrame
print(df.head())
```

```python

df.columns
```

```python
# Crear la columna TIPO
df['TIPO'] = df['ESTADO'].apply(lambda x: 'Entrada' if x == 'Enviado' else 'Salida')
```

```python
df['FECHA_REGISTRO_HH'] = pd.to_datetime(df['FECHA_REGISTRO_HH']) 
```

```python
# Filtrar para obtener solo las fechas de salida
df_salida = df[df['TIPO'] == 'Salida'][['DOCO_ID', 'FECHA_REGISTRO_HH']].rename(columns={'FECHA_REGISTRO_HH': 'fecha_salida'})

# Mostrar el DataFrame de salidas
print("DataFrame de Salidas:")
print(df_salida)
```

```python
# Filtrar el DataFrame original para obtener solo las entradas
df_entrada = df[df['TIPO'] == 'Entrada']
```

```python
# Unir el DataFrame de salidas con el DataFrame de entradas por ID
df_merged = pd.merge(df_entrada, df_salida, on='DOCO_ID', how='left')
```

```python
df_merged
```

```python
# Calcular la diferencia en horas entre fecha_salida y fecha (entrada)
df_merged['diferencia_horas'] = (df_merged['fecha_salida'] - df_merged['FECHA_REGISTRO_HH']).dt.total_seconds() / 3600
```

```python
df_merged
```

```python
# Contar cuántos NaT hay en la columna 'fecha_salida'
num_nats = df_merged['fecha_salida'].isna().sum()
num_nats
```

```python
# Contar cuántos NaT hay en la columna 'fecha_salida'
num_not = df_merged['fecha_salida'].notna().sum()
num_not
```

```python
# Definir el umbral de 48 horas
umbral = 48

# Calcular la desviación respecto al umbral
df_merged['desviacion'] = df_merged['diferencia_horas'] - umbral

# Determinar el estado teniendo en cuenta los casos sin horas de salida
df_merged['estado'] = df_merged.apply(
    lambda row: 'Sin gestionar' if pd.isna(row['diferencia_horas']) 
    else ('Fuera de plazo' if row['desviacion'] > 0 
          else ('En plazo' if row['desviacion'] < 0 
                else 'Igual')), 
    axis=1
)
```

```python
from datetime import datetime
```

```python
# Obtener el mes actual y los tres meses anteriores
hoy = datetime.now()
mes_actual = hoy.month
año_actual = hoy.year

# Crear una lista de los tres meses anteriores y el mes actual
meses_validos = [(mes_actual - i - 1) % 12 + 1 for i in range(4)]
año_valido = año_actual if mes_actual in meses_validos else año_actual - 1

# Función para obtener la etiqueta del mes
def etiqueta_mes(fecha):
    mes = fecha.month
    año = fecha.year
    
    if mes == mes_actual and año == año_actual:
        return f"{hoy.strftime('%B')} {año_actual}"
    
    if mes in meses_validos:
        return f"{fecha.strftime('%B')} {año}"
    
    return 'Fuera de rango'

# Aplicar la función al DataFrame
df_merged['etiqueta_mes'] = df_merged['FECHA_REGISTRO_HH'].apply(etiqueta_mes)

print(df_merged)
```
